## 데이터 수집 진행 상황 관리

중단된 지점부터 재개할 수 있도록 진행 상황을 저장합니다.

```python
import soccerdata as sd

fb = sd.FBref("ENG-Premier League", "2018")  # 2017/18 시즌
player_match_stats = fb.read_player_match_stats(stat_type="summary")  # summary로 변경

# 사용 가능한 stat_type 옵션:
# 'summary', 'keepers', 'passing', 'passing_types', 'defense', 'possession', 'misc'
```

```python
# 데이터 확인
print(f"데이터 크기: {player_match_stats.shape}")
print(f"\n컬럼: {list(player_match_stats.columns)}")
player_match_stats.head()
```

```python
# CSV 파일로 저장
output_path = "../data/player_match_stats_2018.csv"
player_match_stats.to_csv(output_path, index=True, encoding='utf-8-sig')
print(f"저장 완료: {output_path}")
```

## 데이터 필터링 및 조회

다양한 조건으로 경기 데이터를 필터링하여 확인할 수 있습니다.

```python
import pandas as pd

# 5. 특정 경기의 모든 선수 데이터 보기 (홈팀 vs 원정팀)
home_team = "Liverpool"
away_team = "Manchester City"

match_data = player_match_stats[
    (player_match_stats.index.get_level_values('team').str.contains(home_team, case=False, na=False)) |
    (player_match_stats.index.get_level_values('team').str.contains(away_team, case=False, na=False))
]

# 같은 날짜에 두 팀이 모두 있는 경기 찾기
if len(match_data) > 0:
    print(f"\n{home_team} vs {away_team} 경기:")
    match_data.head(30)
else:
    print(f"{home_team}와 {away_team}의 경기를 찾을 수 없습니다.")
```

```python
# 6. 데이터 요약 통계 - 선수별 평균 성적

player_avg = player_match_stats.groupby('player').agg({
    'goals': 'sum',
    'shots': 'sum', 
    'minutes': 'sum'
}).sort_values('goals', ascending=False)

print("\n선수별 시즌 통산 기록 (상위 20명):")
player_avg.head(20)
```
